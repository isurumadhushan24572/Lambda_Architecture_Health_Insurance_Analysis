```python
# Import necessary libraries
import asyncio
import json
import random
from datetime import datetime
from azure.eventhub import EventData
from azure.eventhub.aio import EventHubProducerClient

# Event Hub details
EVENT_HUB_CONNECTION_STR = "Endpoint=sb://<your-event-hub-namespace>.servicebus.windows.net/;SharedAccessKeyName=<your-policy-name>;SharedAccessKey=<your"
EVENT_HUB_NAME = "<your-event-hub-name>"

# Base sample to randomize
base_record = {
    "Referring_NPI": 1073627059,
    "Referring_Provider_Last_Name_/_Organization_Name": "TESTING",
    "Referring_Provider_First_Name": "IMAGING",
    "Referring_Provider_Middle_Initial": "N",
    "Referring_Credentials": "MD",
    "Referring_Entity_Code": "I",
    "Referring_Provider_Street_1": "12345 ANYWHERE BLVD",
    "Referring_Provider_Street_2": "",
    "Referring_Provider_City": "ST LOUIS",
    "Referring_Provider_State": "MO",
    "Referring_Provider_State_FIPS_Code": "29",
    "Referring_Provider_ZIP": "63141",
    "Rfrg_Prvdr_RUCA_Cat": "Metropolitan area core",
    "Referring_Provider_RUCA": 1.0,
    "Referring_Provider_RUCA_Description": "Metropolitan area core: primary flow within an urbanized area (UA)",
    "Referring_Provider_Country": "US",
    "Referring_Provider_Specialty_Code": 30.0,
    "Referring_Provider_Specialty_Description": "Diagnostic Radiology",
    "Referring_Provider_Specialty_Source": "PECOS",
    "RBCS_Level": "A",
    "RBCS_ID": "1",
    "RBCS_Description": "Standard Imaging Services",
    "HCPCS_Code": "71020",
    "HCPCS_Description": "Chest x-ray 2vw frontal&latl",
    "Supplier_Rental_Indicator": "N",
    "Number_of_Suppliers": 3945.0,
    "Number_of_Supplier_Beneficiaries": 326844.0,
    "Number_of_Supplier_Claims": 462232.0,
    "Number_of_Supplier_Services": 468822.0,
    "Average_Supplier_Submitted_Charges": 28.23,
    "Average_Supplier_Medicare_Allowed_Amount": 10.54,
    "Average_Supplier_Medicare_Payment_Amount": 8.62,
    "Average_Supplier_Medicare_Standard_Payment_Amount": 8.61
}

def generate_synthetic_record():
    record = base_record.copy()
    record["Referring_NPI"] = random.randint(1000000000, 9999999999)
    record["Number_of_Suppliers"] = round(random.uniform(1, 100), 1)
    record["Number_of_Supplier_Beneficiaries"] = round(random.uniform(1, 100), 1)
    record["Number_of_Supplier_Claims"] = round(random.uniform(1, 200), 1)
    record["Number_of_Supplier_Services"] = round(random.uniform(1, 200), 1)
    record["Average_Supplier_Submitted_Charges"] = round(random.uniform(20, 40), 2)
    record["Average_Supplier_Medicare_Allowed_Amount"] = round(random.uniform(8, 15), 2)
    record["Average_Supplier_Medicare_Payment_Amount"] = round(random.uniform(5, 12), 2)
    record["Average_Supplier_Medicare_Standard_Payment_Amount"] = round(random.uniform(5, 12), 2)
    # record["timestamp"] = datetime.utcnow().isoformat() + "Z"
    return record

async def send_realtime_events():
    producer = EventHubProducerClient.from_connection_string(
        conn_str=EVENT_HUB_CONNECTION_STR,
        eventhub_name=EVENT_HUB_NAME
    )

    async with producer:
        while True:
            record = generate_synthetic_record()
            event_body = json.dumps(record)

            event_data_batch = await producer.create_batch()
            event_data_batch.add(EventData(event_body))
            await producer.send_batch(event_data_batch)

            print(f"NPI {record['Referring_NPI']}")

            await asyncio.sleep(2)  # Wait 30 seconds between sends



# In Jupyter/Notebook use:
await send_realtime_events()
```
